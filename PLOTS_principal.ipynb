```python
#PLOT SHOWING THE VARIATIONS WITH RESPECT TO THE AMOUNT OF SECONDS USED FOR THE MVSTD

list_cones=[]
list_clocks=[]


for i in np.arange(10,600):

    rolling1=coness_['col'].rolling(i,center=True).std()
    rolling2=clockss_['col'].rolling(i,center=True).std()


    indexx1=table1['Index1'][0]
    indexx2=table1['Index2'][0]

    coness = rolling1.loc[indexx1-15:indexx2+15]
    clockss = rolling2.loc[indexx1-15:indexx2+15]
    list_cones.append(np.nanmax(coness))
    list_clocks.append(np.nanmax(clockss))

plt.figure()    
plt.plot(np.arange(10,600),list_cones,label='Cones angles')
plt.plot(np.arange(10,600),list_clocks,label='Clocks angles')
plt.xlabel('Amount of seconds used for the mstd (s)')
plt.ylabel('angles (°)')
plt.title('Single event (August 2015)')
plt.legend()
```
